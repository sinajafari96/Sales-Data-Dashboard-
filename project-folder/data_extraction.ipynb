# seprate dim and fact tabels

to seprate dim and fact tables we should first define some keys to be able to relate tables to each other after sepration

```python
Sales_dataframe['RegionKey'] = pd.factorize(Sales_dataframe['Region'])[0] + 1
Sales_dataframe['ProductKey'] = pd.factorize(Sales_dataframe['ProductName'])[0] + 1
Sales_dataframe['SubcategoryKey'] = pd.factorize(Sales_dataframe['SubcategoryName'])[0] + 1
Sales_dataframe['CategoryKey'] = pd.factorize(Sales_dataframe['CategoryName'])[0] + 1
Sales_dataframe['CustomerKey'] = pd.factorize(Sales_dataframe['EmailAddress'])[0] + 1

```

```python
Sales_dataframe.columns
```

now dim tables will be created for those values that are unique 

```python
products = Sales_dataframe[['ProductKey', 'SubcategoryKey', 'ProductSKU', 'ProductName', 
                      'ModelName', 'ProductDescription', 'ProductColor', 'ProductSize', 
                      'ProductStyle', 'ProductCost', 'ProductPrice']]

products = products.drop_duplicates(subset=['ProductKey'], keep='first')


Customers = Sales_dataframe[['CustomerKey', 'Prefix', 'FirstName', 'LastName', 
                      'BirthDate', 'MaritalStatus', 'Gender', 'EmailAddress', 
                      'AnnualIncome', 'TotalChildren', 'EducationLevel', 'Occupation', 'HomeOwner']]

Customers = Customers.drop_duplicates(subset=['CustomerKey'], keep='first')


Products_Subcategory = Sales_dataframe[['SubcategoryKey', 'CategoryKey', 'SubcategoryName']]

Products_Subcategory = Products_Subcategory.drop_duplicates(subset=['SubcategoryKey'], keep='first')

Products_Category = Sales_dataframe[['CategoryKey', 'CategoryName']]

Products_Category = Products_Category.drop_duplicates(subset=['CategoryKey'], keep='first')

Regions = Sales_dataframe[['RegionKey', 'Region', 'Country', 'Continent']]

Regions = Regions.drop_duplicates(subset=['RegionKey'], keep='first')


date_range = pd.date_range(start='2020-01-01', end='2023-01-01')

Calendar = pd.DataFrame({'Date': date_range.astype(str)})
```

now fact tables will be created for those values that are not unique and includong numbers

```python
Sales_Data = Sales_dataframe[['OrderDate', 'StockDate', 'OrderNumber', 'OrderLineItem', 
                      'OrderQuantity', 'RegionKey', 'ProductKey', 'CustomerKey']]

Sales_Data = Sales_Data.drop_duplicates(subset=['OrderNumber', 'StockDate', 'OrderLineItem'], keep='first')


Returns = Sales_dataframe[['RegionKey', 'ProductKey', 'ReturnDate', 'ReturnQuantity']]

Returns = Returns.drop_duplicates(subset=['RegionKey', 'ProductKey', 'ReturnDate'], keep='first')

Returns.dropna(inplace=True)
```
